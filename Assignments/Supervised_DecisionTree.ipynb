```python
import seaborn as sns
```

```python
df=sns.load_dataset('iris')
```

```python
df.head(4)
```

```python
X = df.drop(['species'], axis=1)
y = df['species']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=.33,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
treemodel=DecisionTreeClassifier()
treemodel.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treemodel,filled= True)
```

We can see only binary division, means this is CART division, Check notes if u forgot. this is post pruning. as till the time the GI become 0, tree kept expanding.... This means overfitting of data, which will definetely have 100% score on train data, but relatively low score on test data because it has captured noise. For that we have to go for limitation of depth of tree.

```python
# prediction for this
y_pred = treemodel.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report, accuracy_score
score = accuracy_score(y_pred,y_test
                      )
score
```

```python
score1 = classification_report(y_pred,y_test)
score1
```
