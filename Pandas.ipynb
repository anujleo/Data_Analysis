```python
# another way of creating datatype

df7 = pd.DataFrame({"customer":['101','102','103','104'], 
                    'category': ['cat2','cat2','cat1','cat3'],
                    'important': ['yes','no','yes','yes'],
                    'sales': [123,52,214,663]},index=[0,1,2,3])

df8 = pd.DataFrame({"customer":['101','103','104','105'], 
                    'color': ['yellow','green','green','blue'],
                    'distance': [12,9,44,21],
                    'sales': [123,214,663,331]},index=[4,5,6,7])
```

```python
df7
```

```python
df8
```

```python
# concaat as on merged line by line column
pd.concat([df7, df8], axis =1 , sort = False)
```

```python
# concat as same as above but sorted
pd.concat([df7, df8], axis = 0, sort = True)
```

```python
# concats side by side as rows
pd.concat([df7, df8], axis = 1, sort = False)
```

```python
df7
```

```python
df8
```

```python
# merger krega customer se aur jitne mutual values hongi likhega jaha nhi hogi vaha NaN likh kr move on.
pd.merge(df7,df8,how='outer', on='customer') #outer merge is union of on
```

```python
#  mergee krega customer se aur jitne mutual values hongi likhega jinme ek ya usse jada common nhi hai usse skip krega pura.
# inner merge is intersection of on
pd.merge(df7,df8,how='inner', on='customer')
```

```python
# SQL k left join vala concept, left  ka sara data but right ka koi insect hora hoga to likh dega otherwise NaN.
# change to right join, it will work.
pd.merge(df7,df8,how='left', on='customer')
```

```python
df9 = pd.DataFrame({'Q1': [101,102,103],
                    'Q2': [201,202,203]},
                   index=['I0','I1','I2'])

df10 = pd.DataFrame({'Q3': [301,302,303],
                    'Q4': [401,402,403]},
                   index=['I0','I2','I3']) 
```

```python


# join behaves just like merge, 
# except instead of using the values of one of the columns 
# to combine data frames, it uses the index labels
df9.join(df10,how='right')
# here it pulls off same result as right join, u can also work for left, outer, inner etc
# outer, inner, left, and right work the same as merge
```

```python
df9.join(df10,how='outer')
# brings out every row 
```

```python
df9.join(df10,how='left')
```

```python
# if some series has multiple of the same value then we can group all the unique entries together
mydict = {'customer': ['Customer 1','Customer 1','Customer2','Customer2','Customer3','Customer3'], 
          'product1': [1.1,2.1,3.8,4.2,5.5,6.9],
          'product2': [8.2,9.1,11.1,5.2,44.66,983]}
df6 = pd.DataFrame(mydict,index=['Purchase 1','Purchase 2','Purchase 3','Purchase 4','Purchase 5','Purchase 6'])
df6
```

```python
grouped_data = df6.groupby('product1')
print(grouped_data)
```

```python
grouped_data.describe()
```

```python
grouped_data = df6.groupby('customer')
print(grouped_data)
```
