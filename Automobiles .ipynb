```python
from sklearn.linear_model import LinearRegression
```

```python
df.tail()
```

```python
df.highway_mpg.max()
```

```python
X = df['highway_mpg'].tolist()
X2 = df['price'].tolist()
k = np.arange(90, 300, 10)
k
```

```python
X = np.array(X).reshape(-1, 1) 
X2 = np.array(X2)
model = LinearRegression()
model.fit(X, X2)
new_X = [60,70,80,90,  90, 100, 110, 120, 130, 140, 150, 160, 170, 180, 190, 200, 210,
       220, 230, 240, 250, 260, 270, 280, 290]
new_X = np.array(new_X).reshape(-1, 1)  # Reshape new_X to a column vector
predictions = model.predict(new_X)







# Plotting the predictions as a line chart
plt.plot(new_X, predictions, marker='o', linestyle='-', label='Predictions')
plt.xlabel('highway_mpg')
plt.ylabel('price')
plt.title('Regression Predictions')
plt.legend()
plt.show()
```
